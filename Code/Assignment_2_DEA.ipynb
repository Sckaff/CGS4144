## Differential expression analysis
Get the pickled data files [here](https://drive.google.com/drive/folders/1jAht30tLDV-gHjQzq18bnwDLYmCiInNU)

```r
# Install the libraries
install.packages("reticulate")
install.packages("BiocManager")
BiocManager::install("DESeq2")
install.packages("ggplot2")
```

```r
# Import library for getting pickle data into R dataframes
library(reticulate)
pd<-import("pandas")
```

```r
# Read pickled counts matrix into R data frame
count_matrix <- pd$read_pickle("../Data/BCCA_Counts_Data_To_Maria.pickle")
count_matrix
```

```r
# Read pickled data labels into R data frame
char_data <-pd$read_pickle("../Data/Disease_Column_To_Maria.pickle")
char_data
```

```r
# Load libraries for performing differential analysis 
library(DESeq2)
library(ggplot2)

# Filter the count matrix to exclude counts below a minimum
filtered_count <- count_matrix %>%
    dplyr::filter(rowSums(.) >= 20)

# Round off the count values
round_count <- round(filtered_count)
```
